### Importing the data

```python
# Open loaded data
swmt_shelf_ave_C_pan005 = xr.open_dataarray('/g/data/g40/kc5856/access_panan_ssp126_SWMT/old_Pan005_SWMT_shelf_rho0/plots/swmt_shelf_sum_ave_C_pan005.nc')
swmt_shelf_ave_M_pan005 = xr.open_dataarray('/g/data/g40/kc5856/access_panan_ssp126_SWMT/old_Pan005_SWMT_shelf_rho0/plots/swmt_shelf_sum_ave_M_pan005.nc')

swmt_shelf_ave_C_pan01 = xr.open_dataarray('/g/data/g40/kc5856/access_panan_ssp126_SWMT/old_Pan01_SWMT_shelf_rho0/plots/swmt_shelf_sum_ave_rho0_C_pan01.nc')
swmt_shelf_ave_M_pan01 = xr.open_dataarray('/g/data/g40/kc5856/access_panan_ssp126_SWMT/old_Pan01_SWMT_shelf_rho0/plots/swmt_shelf_sum_ave_rho0_M_pan01.nc')
```

```python
swmt_shelf_ave_anom_pan005 = swmt_shelf_ave_M_pan005 - swmt_shelf_ave_C_pan005
swmt_shelf_ave_anom_pan01 = swmt_shelf_ave_M_pan01 - swmt_shelf_ave_C_pan01
```

```python
cont_d25 = 27.89
cont_d75 = 27.73
```

```python
fig, axs = plt.subplots(1, 3, figsize=(13.5, 5.33), sharey=True)  # 3 plots side by side

# Subplot 1 (left)
axs[0].plot(swmt_shelf_ave_C_pan01, swmt_shelf_ave_C_pan01['isopycnal_bins'], color='k', label='Control: 3.67 Sv')
axs[0].plot(swmt_shelf_ave_M_pan01, swmt_shelf_ave_M_pan01['isopycnal_bins'], color='r', label='Meltwater: 3.10 Sv')
axs[0].plot([0, 0], [26.5, 28.49], 'k', linewidth=0.5)
axs[0].plot([-5, 15], [cont_d25, cont_d25], 'k--', linewidth=1)
axs[0].plot([-5, 15], [cont_d75, cont_d75], 'k--', linewidth=1)
axs[0].set_xlim((-1.5, 10))
axs[0].set_ylim((28.45, 26.5))
axs[0].set_xlabel('Surface water-mass transformation (Sv)')
axs[0].set_ylabel('Density $\sigma_0$ (kg m$^{-3}$)')
axs[0].legend()

# Subplot 2 (middle)
axs[1].plot(swmt_shelf_ave_C_pan005, swmt_shelf_ave_C_pan005['isopycnal_bins'], color='k', label='Control: 3.74 Sv')
axs[1].plot(swmt_shelf_ave_M_pan005, swmt_shelf_ave_M_pan005['isopycnal_bins'], color='r', label='Meltwater: 3.17 Sv')
axs[1].plot([0, 0], [26.5, 28.49], 'k', linewidth=0.5)
axs[1].plot([-5, 15], [cont_d25, cont_d25], 'k--', linewidth=1)
axs[1].plot([-5, 15], [cont_d75, cont_d75], 'k--', linewidth=1)
axs[1].set_xlim((-1.5, 10))
axs[1].set_xlabel('Surface water-mass transformation (Sv)')
axs[1].legend()

# Subplot 3 (right)
axs[2].plot(swmt_shelf_ave_anom_pan01, swmt_shelf_ave_anom_pan01['isopycnal_bins'], label='Panan 1/10th')
axs[2].plot(swmt_shelf_ave_anom_pan005, swmt_shelf_ave_anom_pan005['isopycnal_bins'], label='Panan 1/20th')
axs[2].plot([0, 0], [26.5, 28.49], 'k', linewidth=0.5)
axs[2].set_xlim((-1.5, 1.5))
axs[2].set_xlabel('Surface water-mass transformation anomaly (Sv)')
axs[2].legend()

# Set y-axis on the right for all and tighten layout
for ax in axs:
    ax.yaxis.set_label_position("left")
    ax.yaxis.tick_left()

plt.tight_layout()
```
