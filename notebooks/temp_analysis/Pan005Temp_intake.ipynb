```python
#Importing data

control = intake.open_esm_datastore('/home/272/kc5856/catalog_dir/panant_005_zstar_access_yr2.json', columns_with_iterables=['variable'])
perturbation = intake.open_esm_datastore('/home/272/kc5856/panantrial/panant_005_zstar_MWonly.json', columns_with_iterables=['variable'])
```

```python
shelf_mask = xr.open_dataset('/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_005deg.nc')['contour_masked_above']
area = control.search(variable='areacello', path=[".*output021.*"]).to_dask()['areacello']
```

```python
def preprocess(cont):
    return cont.sel(yh=shelf_mask['yh'], z_l=slice(None,500))


cont = control.search(
    variable=["thetao", "volcello"],
    frequency="1mon",
    filename=".*ocean_month_z",
    start_date="^2.*" # 2000s+ - this will be easier/nicer when date-based searching is implemented
).to_dask(
    xarray_open_kwargs=dict(
        chunks="auto"
    ),
    preprocess=preprocess
)
```

```python
def preprocess(melt):
    return melt.sel(yh=shelf_mask['yh'], z_l=slice(None,500))


melt = perturbation.search(
    variable=["thetao", "volcello"],
    frequency="1mon",
    filename=".*ocean_month_z",
    start_date="^2.*" # 2000s+ - this will be easier/nicer when date-based searching is implemented
).to_dask(
    xarray_open_kwargs=dict(
        chunks="auto"
    ),
    preprocess=preprocess
)
```

```python
tempC = cont['thetao']
volcC = cont['volcello']
thikC = cont['volcello']/area
```

```python
tempM = melt['thetao']
volcM = melt['volcello']
thikM = melt['volcello']/area
```

```python
temp_ave_melt = tempM.weighted(thikM.fillna(0)).mean('z_l').where(shelf_mask == 0).weighted(area.fillna(0)).mean(['xh', 'yh'])
temp_ave_cont = tempC.weighted(thikC.fillna(0)).mean('z_l').where(shelf_mask == 0).weighted(area.fillna(0)).mean(['xh', 'yh'])
```
