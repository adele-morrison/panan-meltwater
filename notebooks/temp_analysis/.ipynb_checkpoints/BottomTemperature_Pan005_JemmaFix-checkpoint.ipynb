```python
## Creating shelf mask another way

def shelf_mask_isobath(var, model_dict):

    paths = {
             "mom5": "/g/data/ik11/grids/Antarctic_slope_contour_1000m.npz",
             "mom6_01": "/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_01deg.nc",
            "mom6_005": "/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_005deg.nc"
             }

    var = var.cf.sel({'latitude': slice(-90, -59)})

    if paths[model_dict][-3:] == '.nc':
        shelf_mask = xr.open_dataset(paths[model_dict])['contour_masked_above']
    else:
        contour_file = np.load(paths[model_dict])
        shelf_mask = xr.DataArray(contour_file['contour_masked_above'],
                                  coords = var.coords, 
                                  dims = var.dims,
                                  name = 'contour_masked_above')
    
    shelf_mask = xr.where(shelf_mask == 0, 1, 0)
    masked_var = var * shelf_mask
    
    return masked_var, shelf_mask
```

```python
depth = xr.open_dataset('/g/data/ol01/outputs/mom6-panan/panant-005-zstar-ACCESSyr2/output239/20101201.ocean_static.nc')['deptho']
land_mask = (0 * depth).fillna(1)
depth_shelf, shelf_mask = shelf_mask_isobath(depth, 'mom6_005')
```

```python
#Time variant variables
def preprocess(ds):
    ds = ds[['thetao', 'volcello','so']]# .sel(yh=shelf_mask['yh']) is taking ages, we can trim later
    return ds

cont = xr.open_mfdataset(pathsC,
                         preprocess = preprocess,
                         #chunks='auto',
                         chunks = {'time':1,'z_l':38,'yh':564,'xh':600},
                         decode_times=True,
                         decode_timedelta=False,
                         parallel=True,
                         # compat='override',
                         # coords='minimal',
                        )
```

```python
cont
```

```python
# Gonna make the bold assumption that the bottom doesn't move in time
bottom_cell = cont.z_l.where(np.isfinite(cont.thetao.isel(time=0))) # mask out depth where there's ocean
all_nan = (bottom_cell.sum('z_l')==0) # any lat/lon that don't have ocean
#del all_nan['z_l'] # make it happy to broadcast?
bottom_cell = bottom_cell.where(~all_nan,0) # Just fill these with zeros for the moment, because there's no ocean data later so it'll end up as a NaN anyway
bottom_cell = bottom_cell.argmax('z_l').squeeze() # Find max depth, drop time dimension with squeeze
del bottom_cell['time'] # Otherwise xarray tries to be too careful and gets upset about times not always matching
bottom_cell.load() # Pretty small, easy to load, we only want this once
bottom_cell.plot()
```

```python
bottom_cell
```

```python
bottom_cont = cont.isel(z_l=bottom_cell)
bottom_cont
```

```python
# Calculating pressure
p_cont = gsw.p_from_z(-bottom_cont['z_l'], bottom_cont['yh'])
p_cont
```

```python
# Calculating absolute salinity
SA_cont = gsw.SA_from_SP(bottom_cont['so'], p_cont, bottom_cont['xh'], bottom_cont['yh'])
SA_cont
```

```python
# Calculating conservative temperature
tempC = gsw.CT_from_pt(SA_cont,bottom_cont['thetao'])

# You may want to put some better names in here, but it was previously self-identifying as salinity
tempC = tempC.rename('cons_t')
tempC.attrs['long_name'] = 'Conservative Temperature'
tempC.attrs['units'] = 'deg C'
tempC.attrs['standard_name'] = 'conservative_temperature'

tempC
```

```python
masked_tempC = tempC.where(shelf_mask,drop=True)
```

```python
masked_tempC.load() # 20 min on xx-large
```
