## Heat budget calculation

We assume a closed heat budget of the following form:
$$ \text{surface heat flux} + \text{cross-shelf heat flux} + \text{along shelf heat flux} = 0 $$  
Surface heat flux is given by:
- `hfds` = surface_downward_heat_flux_in_sea_water
- Surface ocean heat flux from SW+LW+latent+sensible+masstransfer+frazil+seaice_melt_heat
  
Along shelf heat flux is found by calculating heat flux through longitudinal transects:
- `temp_xflux_adv` = `cp0` * `rhopot0` * `dzt` * `dyt` * `uo` * `thetao`
- use 0 or 2 for rho? Mass transport is in 2
  
Cross-shelf heat flux is found by:
- $\text{cross shelf flux} = - \text{surface flux} - \text{along shelf flux}$
