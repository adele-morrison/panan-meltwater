```python
import glob

import matplotlib.pyplot as plt
import xarray as xr
import cf_xarray as cfxr
import numpy as np

import cartopy.feature as cft
import matplotlib.path as mpath

import warnings # ignore these warnings
warnings.filterwarnings("ignore", category = FutureWarning)
warnings.filterwarnings("ignore", category = UserWarning)
warnings.filterwarnings("ignore", category = RuntimeWarning)
```

```python
#Starting a dask client
from os import environ
environ["PYTHONWARNINGS"] = "ignore"

from dask.distributed import Client
client = Client(threads_per_worker=1)
client
```

```python
# Importing data
pathsC = np.sort(glob.glob('/g/data/ol01/outputs/mom6-panan/panant-005-zstar-ACCESSyr2/outpu*/*.ocean_month.nc'))
pathsM = np.sort(glob.glob('/g/data/g40/akm157/model_output/mom6-panan/panant-005-zstar-ssp126-MW-only/outpu*/*.ocean_month.nc'))[:60]
```

```python
# Time variant variables
def preprocess(ds):
    ds = ds['wfo']
    return ds

cont = xr.open_mfdataset(pathsC, preprocess = preprocess, chunks = 'auto')
melt = xr.open_mfdataset(pathsM, preprocess = preprocess, chunks = 'auto')
```

```python
# Taking time mean
WFOpert = melt['wfo'].sel(time=slice('2001-01','2010-12')).mean(dim='time')
WFOcont = cont['wfo'].sel(time=slice('2001-01','2010-12')).mean(dim='time')

# Anomaly
WFOanom = WFOpert - WFOcont
```

```python
WFOanom = WFOanom.load()
```

```python
plt.figure(figsize=(18, 5))
WFOanom.plot()
plt.ylim(-80,-65)
```

```python
plt.figure(figsize=(7, 8))
WFOanom.plot()
plt.ylim(-70,-69)
plt.xlim(-73,-71)
```
