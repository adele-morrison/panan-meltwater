```python
def shelf_mask_isobath_MOM5(var):
    '''
    Masks ACCESS-OM2-01 variables by the region polewards of the 1000m isobath as computed using 
    a script contributed by Adele Morrison.
    Only to be used with ACCESS-OM2-0.1 output!
    '''
    contour_file = np.load('/g/data/ik11/grids/Antarctic_slope_contour_1000m.npz')
    
    shelf_mask = contour_file['contour_masked_above']
    yt_ocean = contour_file['yt_ocean']
    xt_ocean = contour_file['xt_ocean']
    
    # in this file the points along the isobath are given a positive value, the points outside (northwards) 
    # of the isobath are given a value of -100 and all the points on the continental shelf have a value of 0 
    # so we mask for the 0 values 
    shelf_mask[np.where(shelf_mask!=0)] = np.nan
    shelf_mask = shelf_mask+1
    shelf_map  = np.nan_to_num(shelf_mask)
    shelf_mask = xr.DataArray(shelf_mask, coords = [('yt_ocean', yt_ocean), ('xt_ocean', xt_ocean)])
    shelf_map  = xr.DataArray(shelf_map,  coords = [('yt_ocean', yt_ocean), ('xt_ocean', xt_ocean)])
    
    # then we want to multiply the variable with the mask so we need to account for the shape of the mask. 
    # The mask uses a northern cutoff of 59S.
    masked_var = var.sel(yt_ocean = slice(-90, -59.03)) * shelf_mask

    return masked_var, shelf_map

def shelf_mask_isobathMOM6(var, contour_depth, resolution, output_mask=False):
    '''
    Masks varibales by the region polewards of a given isobath
    '''
    
    ds_contour = xr.open_dataset(
        '/home/142/cs6673/work/mom6_comparison/Antarctic_slope_contours/' +
        'Antarctic_slope_contour_' + str(contour_depth) + 'm_MOM6_' +
        resolution + 'deg.nc')
    
    shelf_mask = ds_contour.contour_masked_above.sel(yh=slice(var.yh[0], var.yh[-1]))
    yh = ds_contour.yh.sel(yh=slice(var.yh[0], var.yh[-1]))
    xh = ds_contour.xh
    
    # in this file the points along the isobath are given a positive value, the points outside (northwards) 
    # of the isobath are given a value of -100 and all the points on the continental shelf have a value of 0 
    # so we mask for the 0 values 
    shelf_mask = shelf_mask.where(shelf_mask == 0)+1
    shelf_mask = shelf_mask.where(shelf_mask == 1, 0)
    
    # multiply the variable with the mask
    masked_var = var * shelf_mask
    
    if output_mask == True:
        return masked_var, shelf_mask
    else:
        return masked_var
```

```python
#defining multiline plot
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

def multiline(xs, ys, c, ax=None, **kwargs):
    """Plot lines with different colorings

    Parameters
    ----------
    xs : iterable container of x coordinates
    ys : iterable container of y coordinates
    c : iterable container of numbers mapped to colormap
    ax (optional): Axes to plot on.
    kwargs (optional): passed to LineCollection

    Notes:
        len(xs) == len(ys) == len(c) is the number of line segments
        len(xs[i]) == len(ys[i]) is the number of points for each line (indexed by i)

    Returns
    -------
    lc : LineCollection instance.
    """

    # find axes
    ax = plt.gca() if ax is None else ax

    # create LineCollection
    segments = [np.column_stack([x, y]) for x, y in zip(xs, ys)]
    lc = LineCollection(segments, **kwargs)

    # set coloring of line segments
    #    Note: I get an error if I pass c as a list here... not sure why.
    lc.set_array(np.asarray(c))

    # add lines to axes and rescale 
    #    Note: adding a collection doesn't autoscalee xlim/ylim
    ax.add_collection(lc)
    ax.autoscale()
    return lc
```

Let's try to divide the change in freshwater transport into the velocity and salinity component to see if stronger mass transport in general or lower surface salinities are causing the observed changes.

Let's take for example Equation 1, and redefine the absolute salinity $S$ as $S_{kg}$, in units of of $[kg_{salt}\ / kg_{total}]$ to make it easier, i.e.,  $ S_{kg} =1 - S/1000$. The zonal direction transport becomes:

$$
\begin{equation}
FW_x\ \hat i =  Q_x \times S_{kg} \  \ \ \ [kg\ s^{-1}]
\end{equation}
$$

Now let's redefine the salinity $S_{kg}$ and  $Q_{x}$ again for the MOM6_5m experiment, based on its anomaly from the 1m experiment, as in:

$$
\begin{equation}
S^{5m}_{kg} (t) =  S^{1m}_{kg} + \Delta S^{5m-1m}_{kg}
\end{equation}
$$

$$
\begin{equation}
Q^{5m}_{x} (t) =  Q^{1m}_{x} + \Delta Q^{5m-1m}_{x}
\end{equation}
$$

Now let's apply that formula above to the freshwater transport and do some algebra...


$$
\begin{equation}
FW_x^{5m}\ = (Q^{1m}_x\ +\ \Delta Q^{5m-1m}_{x}) \times (S^{1m}_{kg}\ +\ \Delta S^{5m-1m}_{kg})\ \ 
\end{equation}
$$
... Which after a tiny ammount of algebra gives ...
$$
\begin{equation}
FW_x^{5m}\ = \ \ (Q^{1m}_x \times S^{1m}_{kg}) \ + \ (Q^{1m}_x \times \Delta S^{5m-1m}_{kg}) \ + (\Delta Q^{5m-1m}_{x} \times S^{1m}_{kg}) \ + (\Delta Q^{5m-1m}_{x} \times \Delta S^{5m-1m}_{kg})\ \ \ \ \ \ [Eq.\ 1]
\end{equation}
$$


and since $FW_x^{1m} = (Q^{1m}_x \times S^{1m}_{kg})$, this equation becomes:
$$
\begin{equation}
\Delta FW_x^{5m - 1m}\ = \ \ (Q^{1m}_x \times \Delta S^{5m-1m}_{kg}) \ + (\Delta Q^{5m-1m}_{x} \times S^{1m}_{kg}) \ + (\Delta Q^{5m-1m}_{x} \times \Delta S^{5m-1m}_{kg})\ \ \ \ \ \ [Eq.\ 2]
\end{equation}
$$


This decomposiiton allow us then top separate the difference between 1mtop and 5mtop freshwater flux simulation into a hidrographic changes effect term $I:\ (Q^{1m}_x \times \Delta S^{5m-1m}_{kg})$, transport/circulation changes effect $II:\  (\Delta Q^{5m-1m}_{x} \times S^{1m}_{kg})$ and one rest leftover co-variability effect $III: \ (\Delta Q^{5m-1m}_{x} \times \Delta S^{5m-1m}_{kg})$, and can help us determine if a stronger transport comes from changes in the velocity field or just in the freshwater content at surface


The decompositionn above was done in terms of direction ($x$) transport, bue we can eliminate the effect of the direction and apply it straight to the transport accross contour ($FW_☩$) as in:

$$
\begin{equation}
\Delta FW_☩^{5m - 1m}\ = \ \ (Q^{1m}_☩ \times \Delta S^{5m-1m}_{kg}) \ + (\Delta Q^{5m-1m}_{☩} \times S^{1m}_{kg}) \ + (\Delta Q^{5m-1m}_{☩} \times \Delta S^{5m-1m}_{kg})\ \ \ \ \ \ [Eq.\ 3]
\end{equation}
$$

# Salt budget

The salinity bdget in the model can be defined by:

$$
\begin{equation}
DS/Dt \ =\ F_{adv} \ + F_{sub} \ + F_{river} \ + \ F_{kpp} \ + \ F_{impl} \ + \ F_{surf}\ \ \ \  [kg\ s^{-1}\ m^{-2}]
\end{equation}
$$

Where $DS/Dt$ is the salinity tendency, $F_{adv}$ is the salinity advective convergence, $ F_{sub} $  and $F_{river}$ is the salinity tendency due to submesoscale processes snd river input respectively. $F_{kpp}$ is the salinity trend due to KPP ocean surface boundary layer parameterization,. $F_{impl} $ is the implicit vert diffusion of Practical Salinity. $F_{surf}$ is the salinity flux at surface, which will be treated as the rest of the balance sum.

Online calcualted Surface forcing term $F_{surf}$ should be the sum of *'sfc_salt_flux_ice'* + *'sfc_salt_flux_restore'*

#### Effect ot salt content on Salinity


Salinity can change either due to changes insalt content or freshwtaer content. Let's calculate, using DsDt, how much the salt content change determined the surface freshening in 5mtop

```python
#creating a time duration matrix in which all values are equal to the number of seconds i one day for 1m and 5mtop
Time_Duration_1m = salt_tendency_1m_shelf_surfsum.copy()
Time_Duration_1m.name='Duration_s'
Time_Duration_1m = (((Time_Duration_1m.fillna(0)*0)+1)+(24*60*60))
Time_Duration_1m_12 = (((Time_Duration_1m.copy().fillna(0)*0)+1)+(12*60*60))

Time_Duration_5m = salt_tendency_5m_shelf_surfsum.copy()
Time_Duration_5m.name='Duration_s'
Time_Duration_5m = (((Time_Duration_5m.fillna(0)*0)+1)+(24*60*60))
Time_Duration_5m_12 = (((Time_Duration_5m.copy().fillna(0)*0)+1)+(12*60*60))
```

```python
# importing dailynsalinities
### 1mtop
sal_1m = cc.querying.getvar(expt_1mtop_abs, 'salt', temp_session,\
                                          frequency='1 daily',chunks={'time': '200MB'})\
.rename({'xt_ocean_sub01':'xt_ocean','yt_ocean_sub01':'yt_ocean'}).sel(st_ocean=depth_slice)
sal_1m['yt_ocean']=yt_ocean_slice
sal_1m, shelf_map = shelf_mask_isobath_MOM5(sal_1m) #masking off shelf
sal_1m = sal_1m.weighted((tarea_MOM5*mean_DZ_1m).fillna(0)).mean(('xt_ocean','yt_ocean','st_ocean'))



### 5mtop
sal_5m = cc.querying.getvar(expt_5mtop_abs, 'salt', temp_session,\
                                          frequency='1 daily',chunks={'time': '200MB'})\
.rename({'xt_ocean_sub01':'xt_ocean','yt_ocean_sub01':'yt_ocean'}).sel(st_ocean=depth_slice)
sal_5m['yt_ocean']=yt_ocean_slice
sal_5m, shelf_map = shelf_mask_isobath_MOM5(sal_5m) #masking off shelf
sal_5m = sal_5m.weighted((tarea_MOM5*mean_DZ_5m).fillna(0)).mean(('xt_ocean','yt_ocean','st_ocean'))

sal_1m.load()
sal_5m.load()
```

```python
sigma0_1m = cc.querying.getvar(expt_1mtop_abs, 'pot_rho_0', temp_session,\
                                          frequency='1 daily',chunks={'time': '200MB'},n=1)\
.rename({'xt_ocean_sub01':'xt_ocean','yt_ocean_sub01':'yt_ocean'}).sel(st_ocean=depth_slice)
sigma0_1m['yt_ocean']=yt_ocean_slice
sigma0_1m_shelf, shelf_map = shelf_mask_isobath_MOM5(sigma0_1m) #masking off shelf

# #we will need an initial salt mass to do that. let's stipulate one reference one for 33.5 of saltinity
Imass_1m = ((tarea_shelf_on_ocean*mean_DZ_1m)*sigma0_1m_shelf).sel(st_ocean=depth_slice).isel(time=0).sum()
Isalt_1m = Imass_1m*sal_1m[0]*1e-3
# Isalt = ((tarea_shelf_on_ocean*mean_DZ_1m)*sigma0_1m_shelf).sel(st_ocean=depth_slice).isel(time=0).sum()
Imass_1m.load()
Isalt_1m.load()
```

```python
#this is rather not a salt content itself, but the salt content change, 
#as we dont have the salt contant in the IC



### for 1mtop
#the end point salinity after 24h of fluxes will be
salt_content_end_day = (salt_tendency_1m_shelf_surfsum*Time_Duration_1m).cumsum('time')
#the salt flux in half timestep will be (removing values to get to the middle of the day)
salt_content_middle_day = (salt_tendency_1m_shelf_surfsum*Time_Duration_1m_12)
salt_predicted_by_budget_1m = salt_content_end_day -salt_content_middle_day + Isalt_1m
del salt_content_end_day, salt_content_middle_day

### for 5mtop
#the end point salinity after 24h of fluxes will be
salt_content_end_day = (salt_tendency_5m_shelf_surfsum*Time_Duration_5m).cumsum('time')
#the salt flux in half timestep will be (removing values to get to the middle of the day)
salt_content_middle_day = (salt_tendency_5m_shelf_surfsum*Time_Duration_5m_12)
salt_predicted_by_budget_5m = salt_content_end_day -salt_content_middle_day + Isalt_1m
del salt_content_end_day, salt_content_middle_day
```

To calculate the salinity change due to changes in salt content, we need to calclate also the total ocean mass. We can do that by dividing the salt content by the (salinity*1e-3)

The total mass will be

```python
#total mass will be
mass_1m = salt_predicted_by_budget_1m/(sal_1m*1e-3)
mass_5m = salt_predicted_by_budget_5m/(sal_5m*1e-3)

#The freshwater mass will be
FWmass_1m =  mass_1m - salt_predicted_by_budget_1m
FWmass_5m =  mass_5m - salt_predicted_by_budget_5m
```

That means that the salinties anomalies, due to changes in salt budget are:

```python
#recalcualting the salinity for sanity check
offSalinity_1m = (salt_predicted_by_budget_1m/mass_1m)*1e3
offSalinity_5m = (salt_predicted_by_budget_5m/mass_5m)*1e3

```

```python
#sanity check
(sal_5m-sal_1m).plot(linewidth=3,label='online salinity')
(offSalinity_5m - offSalinity_1m).plot(linestyle = ':',linewidth=3,label = 'offline salinity')
plt.legend()
plt.ylabel('Salinity')
```

lets now calcualte the component of salinity change due to salt change and due to freshwater change

```python
#Salinty changes due to salt content (mass of freshwater is constant)
S_bysalt_1m= ( salt_predicted_by_budget_1m / \
              (FWmass_1m.isel(time=slice(0,31)).mean() + salt_predicted_by_budget_1m) )*1e3
S_bysalt_5m= ( salt_predicted_by_budget_5m / \
              (FWmass_5m.isel(time=slice(0,31)).mean() + salt_predicted_by_budget_5m) )*1e3


#Salinty changes due to freshwater content (mass of salt is constant)
S_byfresh_1m= ( salt_predicted_by_budget_1m.isel(time=slice(0,31)).mean() /\
               (FWmass_1m + salt_predicted_by_budget_1m.isel(time=slice(0,31)).mean()) )*1e3
S_byfresh_5m= ( salt_predicted_by_budget_5m.isel(time=slice(0,31)).mean() / \
               (FWmass_5m + salt_predicted_by_budget_5m.isel(time=slice(0,31)).mean()) )*1e3
```

```python
(sal_5m-sal_1m).plot(linewidth=3,label='online salinity')

(S_bysalt_5m - S_bysalt_1m).plot(linestyle = ':',linewidth=3,label = 'Salinity by varying salt content')
(S_byfresh_5m - S_byfresh_1m).plot(linestyle = ':',linewidth=3,label = 'Salinity by varying freshwater content')

plt.grid()
plt.legend(fontsize=16, bbox_to_anchor=(0.1, -0.2), loc="upper left")
plt.ylabel('Salinity')
```


Through the plot above, we can see that the signal of salinity change actually comes from changes in salt concentration, instead of freshwater concentration. In summary, for some reason, 5mtop concetrate less salt than 1mtop. This is good cause then the salt budget can actually teach us something

Finally, we need to calculate the net freshwater flux at surface. We will use the knowledge from the surface water mass transformation function for that

```python
def Extract_Fluxes(expt, session, start_time, end_time, lat_north = -59, n = None):

    ## getvar all required variables
    SST = cc.querying.getvar(expt, 'temp', session, frequency='1 monthly',chunks={"time": '200MB'}).isel(st_ocean=0) - 273.15 # SST - conservative temperature in K (sheck this is the case for your run)
    SSS_PSU = cc.querying.getvar(expt, 'salt', session, frequency='1 monthly',chunks={"time": '200MB'}).isel(st_ocean=0)      # SSS - practical salinity (not absolute)
    pme_net = cc.querying.getvar(expt, 'pme_net', session, frequency='1 monthly',chunks={"time": '200MB'})       # mass flux of precip - evap + river
    if expt=='ryf_FWFdist': #explanation... the FWF  from sea ice is removed from pme_river in FWFdist, so I have to add it manually
        wfiform = cc.querying.getvar(expt, 'wfiform', session, frequency='1 monthly',chunks={"time": '200MB'})
        pme_river = pme_river + wfiform
    
    # Note that is sfc_salt_flux_ice is not saved, you can use sfc_salt_flux_ice = 0.005 * melt instead
    sfc_salt_flux_ice = cc.querying.getvar(expt, 'sfc_salt_flux_ice', session, frequency='1 monthly',chunks={"time": '200MB'})         # mass flux of salt from ice formation/melt
    sfc_salt_flux_restore = cc.querying.getvar(expt, 'sfc_salt_flux_restore', session, frequency='1 monthly',chunks={"time": '200MB'}) # mass flux of salt from surface salinity restoring

    ## getvar the components of the net surface heat fux instead of the net_surface_heating variable
    sfc_hflux_from_runoff = cc.querying.getvar(expt, 'sfc_hflux_from_runoff', session, frequency='1 monthly',chunks={"time": '200MB'}) # W/m²
    sfc_hflux_coupler = cc.querying.getvar(expt, 'sfc_hflux_coupler', session, frequency='1 monthly',chunks={"time": '200MB'})         # W/m²
    sfc_hflux_pme = cc.querying.getvar(expt, 'sfc_hflux_pme', session, frequency='1 monthly',chunks={"time": '200MB'})                 # W/m²
    frazil_3d_int_z = cc.querying.getvar(expt, 'frazil_3d_int_z', session, frequency='1 monthly',chunks={"time": '200MB'})             # W/m²
    geolon_t = cc.querying.getvar(expt, 'geolon_t', session, n=1)
    geolat_t = cc.querying.getvar(expt, 'geolat_t', session, n=1)
    
    ## slice for time and latitudinal constraints
    time_slice = slice(start_time, end_time)
    lat_slice = slice(-90, lat_north)
    SST = SST.sel(time=time_slice, yt_ocean=lat_slice)
    SSS_PSU = SSS_PSU.sel(time=time_slice, yt_ocean=lat_slice)
    pme_net = pme_net.sel(time=time_slice, yt_ocean=lat_slice)
    sfc_salt_flux_ice = sfc_salt_flux_ice.sel(time=time_slice, yt_ocean=lat_slice)
    sfc_salt_flux_restore = sfc_salt_flux_restore.sel(time=time_slice, yt_ocean=lat_slice)
    sfc_hflux_from_runoff = sfc_hflux_from_runoff.sel(time=time_slice, yt_ocean=lat_slice)
    sfc_hflux_coupler = sfc_hflux_coupler.sel(time=time_slice, yt_ocean=lat_slice)
    sfc_hflux_pme = sfc_hflux_pme.sel(time=time_slice, yt_ocean=lat_slice)
    frazil_3d_int_z = frazil_3d_int_z.sel(time=time_slice, yt_ocean=lat_slice)
    
    lon_t = geolon_t.sel(yt_ocean=lat_slice)
    lat_t = geolat_t.sel(yt_ocean=lat_slice)
    
    ## extract coordinate arrays
    yt_ocean = SST.yt_ocean.values
    xt_ocean = SST.xt_ocean.values
    st_ocean = cc.querying.getvar(expt,'st_ocean',session,n=1).load()
    time_monthly = SST.time.values
    
    ## construct an xarray of days per month (check this is relevant to your run), simple modification if non integer number of years analysed
    start_month = int(start_time[5:7])
    end_month = int(end_time[5:7])
    n_years = int(len(SST.time)/12)
    months_standard_noleap = np.array([31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31])
    if start_month != 1:
        months_offset_noleap = np.append(months_standard_noleap[(start_month-1):],months_standard_noleap[:(start_month-1)])
    else:
        months_offset_noleap = months_standard_noleap
    
    ## now I use some TEOS-10 gsw functions to compute absolute salinity, then potential density fields
    ## these are numpy functions, if you have memory errors this is a good step to check (though I have found
    ## this works on the VDI for 0.1 degree data, might be issues for very long time periods)
    depth = -st_ocean[0].values # st_ocean value of the uppermost cell
    depth_tile = (lat_t*0 + 1) * depth
    pressure = xr.DataArray(p_from_z(depth_tile, lat_t), coords = [yt_ocean, xt_ocean], dims = ['yt_ocean', 'xt_ocean'], name = 'pressure', attrs = {'units':'dbar'})
    
    # convert units to absolute salinity 
    SSS = xr.DataArray(SA_from_SP(SSS_PSU, pressure, lon_t, lat_t), coords = [time_monthly, yt_ocean, xt_ocean], dims = ['time','yt_ocean', 'xt_ocean'], name = 'sea surface salinity', attrs = {'units':'Absolute Salinity (g/kg)'})
    
    ## SST is already saved as conservative temperature in this run, if you are working with an older run with potential 
    ## temperature saved, conversion will be required (make sure you work with C not K)
    # SST = xr.DataArray(CT_from_pt(SSS_AS,SST_PT), coords = [time_monthly, yt_ocean, xt_ocean], dims = ['time', 'yt_ocean', 'xt_ocean'], name = 'sea surface temperature', attrs = {'units':'Conservative Temperature (C)'})
    # compute potential density referenced to 1000dbar (or referenced otherwise, depending on your purpose)
    pot_rho_1 = xr.DataArray(sigma0(SSS, SST), coords = [time_monthly, yt_ocean, xt_ocean], dims = ['time', 'yt_ocean', 'xt_ocean'], name = 'potential density ref 1000dbar', attrs = {'units':'kg/m^3 (-1000 kg/m^3)'})
    pot_rho_1 = pot_rho_1.load()
    
    # Compute salt transformation (no density binning)
    haline_contraction = xr.DataArray(beta(SSS, SST, pressure), coords = [time_monthly, yt_ocean, xt_ocean], dims = ['time','yt_ocean', 'xt_ocean'], name = 'saline contraction coefficient (constant conservative temp)', attrs = {'units':'kg/g'})

    #freshwater flux components expressly spit out by the model...
    Salt_fluxes_into_FWF = -(sfc_salt_flux_ice + sfc_salt_flux_restore) * (1/(SSS*1e-3)) * (1 - (SSS/1000))#kg of Fwater/m²/s
    #vertical freshwater flux will be
    FWF =  pme_net

    return FWF ## helpful for re-loading


```

```python
%%time
start_time00='1900-01-01'
end_time00='1900-06-30'

FWFs_1m_abs = Extract_Fluxes(expt_1mtop_abs, temp_session, start_time00, end_time00)
FWFs_5m_abs = Extract_Fluxes(expt_5mtop_abs, temp_session, start_time00, end_time00)

# masking off shelf
FWFs_1m_abs_shelf, shelf_map = shelf_mask_isobath_MOM5(FWFs_1m_abs)
FWFs_5m_abs_shelf, shelf_map = shelf_mask_isobath_MOM5(FWFs_5m_abs)

#loading
FWFs_1m_abs_shelf.load()
FWFs_5m_abs_shelf.load()
```
