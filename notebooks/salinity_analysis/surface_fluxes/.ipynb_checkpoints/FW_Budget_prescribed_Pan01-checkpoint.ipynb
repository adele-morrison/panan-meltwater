```python
SSS_ctrl = cc.querying.getvar(control, 'so', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,0,:,:].sel(yh=lat_slice)
SSS_melt = cc.querying.getvar(meltwater, 'so', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,0,:,:].sel(yh=lat_slice)
```

```python
# convert salt fluxes to FW equivalent:

# A. Kiss don't think we should add this salt fluxes here; we're comparing them but the surface FW budget is given by:
# PME_RIVER = RIVER + MELT (wfiform+wfimelt) + PME (pme_net - (wfiform+wfimelt)) 
FW_restore_C = -(salt_flux_added_C*1000)/SSS_ctrl
FW_restore_M = -(salt_flux_added_M*1000)/SSS_melt

# He also says we should use "the fixed bulk salinity of sea ice (ice_salt_concentration = 0.005 kg salt / kg ice), not SSS"
SSSi = 1/0.005

FW_ice_C = -(salt_flux_ice_C*1000)/SSSi
FW_ice_M = -(salt_flux_ice_M*1000)/SSSi
```

```python
# topography data for plotting:
ht = cc.querying.getvar(control,'deptho' , master_session,n=1).sel(yh=lat_slice)
land_mask = np.squeeze(ht.values)*0
land_mask[np.isnan(land_mask)] = 1
land_mask = np.where(land_mask==1,land_mask,np.nan)
land_mask_masked = np.ma.masked_where((land_mask==0),land_mask)
# make land go all the way to -90S:
land_mask_lat = ht.yh.values
land_mask_lat[0] = -90

fontsize=15
```

```python
area_old = xr.open_dataset('/g/data/ol01/outputs/mom6-panan/panant-01-zstar-ACCESSyr2/output050/20050501.ocean_static.nc')['areacello']
varr = xr.open_dataset('/g/data/ol01/outputs/mom6-panan/panant-01-zstar-ACCESSyr2/output050/20050501.ocean_month_z.nc')['so']
area_corr = area_old*(varr.isel(time=0,z_l=0)*0 +1)
```

```python
%%time
import numpy.ma as ma

area_t = area_corr.sel(yh=lat_slice)

# function to do annual avg and horiz. integral: ONLY SHELF
def var_tavg_hint_shelf(var2d):
    var = var2d.mean('time') # annual mean
    var_masked = ma.masked_where(ht > 1000, var) # masked offshore 1000m isobath
    var_masked_area = var_masked*area_t # multiply by area
    var_hint = np.nansum(np.nansum(var_masked_area, axis=1), axis=0) # sum along x/y axis

    return var_hint

# function to do annual avg and horiz. integral: GLOBAL
def var_tavg_hint(var2d):
    var = var2d.mean('time') # annual mean
    var_area = var*area_t # multiply by area
    var_hint = np.nansum(np.nansum(var_area, axis=1), axis=0) # sum along x/y axis

    return var_hint
```

```python
%%time

#pme_net_annual_hint_shelf_ctrl = var_tavg_hint_shelf(pme_net_C)
wfo_annual_hint_shelf_ctrl = var_tavg_hint_shelf(wfo_C)
river_annual_hint_shelf_ctrl = var_tavg_hint_shelf(river_C)
melt_annual_hint_shelf_ctrl = var_tavg_hint_shelf(melt_C)
FW_restore_annual_hint_shelf_ctrl = var_tavg_hint_shelf(FW_restore_C)
FW_ice_annual_hint_shelf_ctrl = var_tavg_hint_shelf(FW_ice_C)
```

```python
%%time

#pme_net_annual_hint_shelf_M = var_tavg_hint_shelf(pme_net_M)
wfo_annual_hint_shelf_M = var_tavg_hint_shelf(wfo_M)
river_annual_hint_shelf_M = var_tavg_hint_shelf(river_M)
melt_annual_hint_shelf_M = var_tavg_hint_shelf(melt_M)
FW_restore_annual_hint_shelf_M = var_tavg_hint_shelf(FW_restore_M)
FW_ice_annual_hint_shelf_M = var_tavg_hint_shelf(FW_ice_M)
```
