### Salinity budget

For the salt/FW fluxes I have the following terms:
- `pme_river` = mass flux of precip-evap+river via sbc (liquid, frozen, evaporation)
   - the equivalent of `wfo` from MOM6 
- `pme_net` = precip-evap into ocean (total w/ restore + normalize)
- `river` = mass flux of river (runoff + calving) entering ocean
- `runoff` = mass flux of liquid river runoff entering ocean
- `evap` = mass flux from evaporation/condensation (>0 enters ocean)
- `melt` = water flux transferred with sea ice form/melt (>0 enters ocean)
- `fprec` = snow falling onto ocean (>0 enters ocean)
- `lprec` = liquid precip (including ice melt/form) into ocean (>0 enters ocean)
- `wfimelt` = water into ocean due to ice melt (>0 enters ocean)
- `wfiform` = water out of ocean due to ice form (>0 enters ocean)
- `sfc_salt_flux_restore` = flux from restoring term
- `sfc_salt_flux_ice` = downward_sea_ice_basal_salt_flux
- `sfc_salt_flux_coupler` = flux from the coupler

I am just going to look at the salt fluxes and do the fresh water fluxes separately I think.

Surface budget is given by:

F_{surface} = `sfc_salt_flux_restore` + `sfc_salt_flux_ice`

Total budget is:
$$ DS/Dt = F_{surface} +F_{advection} $$

So I am trying to find $F_{advection}$ by doing:
$$ F_{advection} = DS/Dt - F_{surface} $$

```python
sfc_salt_flux_restore_C = cc.querying.getvar(control, 'sfc_salt_flux_restore', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:15,:,:].sel(yt_ocean=lat_slice)
sfc_salt_flux_ice_C = cc.querying.getvar(control, 'sfc_salt_flux_ice', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:15,:,:].sel(yt_ocean=lat_slice)
net_sfc_salt_C = sfc_salt_flux_restore_C + sfc_salt_flux_ice_C
net_sfc_salt_C = net_sfc_salt_C.sel(time=slice(start_time,end_time))

sfc_salt_flux_restore_C = sfc_salt_flux_restore_C.sel(time=slice(start_time,end_time))
sfc_salt_flux_ice_C = sfc_salt_flux_ice_C.sel(time=slice(start_time,end_time))
```

```python
sfc_salt_flux_restore_M = cc.querying.getvar(meltwater, 'sfc_salt_flux_restore', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:15,:,:].sel(yt_ocean=lat_slice)
sfc_salt_flux_ice_M = cc.querying.getvar(meltwater, 'sfc_salt_flux_ice', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:15,:,:].sel(yt_ocean=lat_slice)

net_sfc_salt_M = sfc_salt_flux_restore_M + sfc_salt_flux_ice_M
net_sfc_salt_M = net_sfc_salt_M.sel(time=slice(start_time,end_time))

sfc_salt_flux_restore_M = sfc_salt_flux_restore_M.sel(time=slice(start_time,end_time))
sfc_salt_flux_ice_M = sfc_salt_flux_ice_M.sel(time=slice(start_time,end_time))
```

```python
area_old = xr.open_dataset('/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/output012/ocean/ocean_grid.nc')['area_t']
varr = xr.open_dataset('/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/output012/ocean/ocean.nc')['salt']
area_corr = area_old*(varr.isel(time=0,st_ocean=0)*0 +1)

area = area_corr
```

```python
import pandas as pd

# assume net_sfc_salt_C.time is a datetime-like coordinate
time = xr.cftime_range(start=str(net_sfc_salt_C.time.values[0]),
                       end=str(net_sfc_salt_C.time.values[-1]),
                       freq='MS')  # if you have monthly means

# --- compute days in each month from actual timestamps ---
days_in_month = xr.DataArray(
    [pd.Period(str(t)).days_in_month for t in net_sfc_salt_C.time.values],
    coords={'time': net_sfc_salt_C.time},
    dims='time'
)

# convert to seconds per month
sec_in_month = days_in_month * 24 * 60 * 60
```

```python
# This will now be in kg integrated over the year.
net_sfc_salt_C = ((net_sfc_salt_C*sec_in_month).sum('time')*area).sum().compute()
sfc_salt_flux_restore_C = ((sfc_salt_flux_restore_C*sec_in_month).sum('time')*area).sum().compute()
sfc_salt_flux_ice_C = ((sfc_salt_flux_ice_C*sec_in_month).sum('time')*area).sum().compute()


print('Net surface salt input over this period =',net_sfc_salt_C.values/1e20,'x 10^20 kg of salt')
print('sfc_salt_flux_restore control =',sfc_salt_flux_restore_C.values/1e20,'x 10^20 kg of salt')
print('sfc_salt_flux_ice control =',sfc_salt_flux_ice_C.values/1e20,'x 10^20 kg of salt')
```

```python
## Need to change the below to be for the salinity
# salt content change:
# We need to calculate this from snapshots from the restarts.

salt_dummy = cc.querying.getvar(control, 'salt', master_session, n=1,frequency='1 monthly')

# note we had to collate these restarts first:
filepath = '/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/restart039/ocean/ocean_temp_salt.res.nc'
restart1t = xr.open_dataset(filepath)
# only 1 time snapshot:
salt_init = restart1t.salt[0,...]
# fix coords:
salt_init.coords['zaxis_1'] = salt_dummy['st_ocean'].values
salt_init.coords['yaxis_1'] = salt_dummy['yt_ocean'].values
salt_init.coords['xaxis_1'] = salt_dummy['xt_ocean'].values
salt_init = salt_init.rename(({'zaxis_1':'st_ocean', 'yaxis_1':'yt_ocean', 'xaxis_1':'xt_ocean'}))
salt_init = salt_init.drop_vars('Time')

filepath = '/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/restart039/ocean/ocean_thickness.res.nc' 

restart1z = xr.open_dataset(filepath)
# only 1 time snapshot:
dzt_init = restart1z.rho_dzt[0,...]
# fix coords:
dzt_init.coords['zaxis_1'] = salt_dummy['st_ocean'].values
dzt_init.coords['yaxis_1'] = salt_dummy['yt_ocean'].values
dzt_init.coords['xaxis_1'] = salt_dummy['xt_ocean'].values
dzt_init = dzt_init.rename(({'zaxis_1':'st_ocean', 'yaxis_1':'yt_ocean', 'xaxis_1':'xt_ocean'}))
dzt_init = dzt_init.drop_vars('Time')

filepath = '/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/restart043/ocean/ocean_temp_salt.res.nc'
restart2t = xr.open_dataset(filepath)
# only 1 time snapshot:
salt_end = restart2t.temp[0,...]
# fix coords:
salt_end.coords['zaxis_1'] = salt_dummy['st_ocean'].values
salt_end.coords['yaxis_1'] = salt_dummy['yt_ocean'].values
salt_end.coords['xaxis_1'] = salt_dummy['xt_ocean'].values
salt_end = salt_end.rename(({'zaxis_1':'st_ocean', 'yaxis_1':'yt_ocean', 'xaxis_1':'xt_ocean'}))
salt_end = salt_end.drop_vars('Time')

filepath = '/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/restart043/ocean/ocean_thickness.res.nc'
restart2z = xr.open_dataset(filepath)
# only 1 time snapshot:
dzt_end = restart2z.rho_dzt[0,...]
# fix coords:
dzt_end.coords['zaxis_1'] = salt_dummy['st_ocean'].values
dzt_end.coords['yaxis_1'] = salt_dummy['yt_ocean'].values
dzt_end.coords['xaxis_1'] = salt_dummy['xt_ocean'].values
dzt_end = dzt_end.rename(({'zaxis_1':'st_ocean', 'yaxis_1':'yt_ocean', 'xaxis_1':'xt_ocean'}))
dzt_end = dzt_end.drop_vars('Time')
```

```python
# compute salt content change:

# First need to convert salt_init and salt_end into absolute salinity. 
### HAVEN"T DONE THIS YET ###

# compute salt content change:
# Total salt content change (kg of salt)
salt_content_change_M = (((salt_end * dzt_end).sum('st_ocean') - (salt_init * dzt_init).sum('st_ocean')) * area).sum()

print('Meltwater Net salt content change over this period =', salt_content_change_M.values/1e20, 'x 10^12 kg of salt')
```
