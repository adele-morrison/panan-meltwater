```python
%%time 
# Make sure these match the available restarts:
start_time='1910-01-01'
end_time='1919-12-31'

lat_slice  = slice(-80,-59)

# load surface heat flux from different components:
pme_net_C = cc.querying.getvar(control, 'pme_net', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
pme_river_C = cc.querying.getvar(control, 'pme_river', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
river_C = cc.querying.getvar(control, 'runoff', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
wfimelt_C = cc.querying.getvar(control, 'wfimelt', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
wfiform_C = cc.querying.getvar(control, 'wfiform', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
melt_C = wfiform_C + wfimelt_C
sfc_salt_flux_restore_C = cc.querying.getvar(control, 'sfc_salt_flux_restore', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
sfc_salt_flux_ice_C = cc.querying.getvar(control, 'sfc_salt_flux_ice', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
fprec_C = cc.querying.getvar(control, 'fprec', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
lprec_C = cc.querying.getvar(control, 'lprec', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
evap_C = cc.querying.getvar(control, 'evap', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
```

```python
%%time
# load surface heat flux from different components:
pme_net_M = cc.querying.getvar(meltwater, 'pme_net', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
pme_river_M = cc.querying.getvar(meltwater, 'pme_river', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
river_M = cc.querying.getvar(meltwater, 'runoff', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
wfimelt_M = cc.querying.getvar(meltwater, 'wfimelt', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
wfiform_M = cc.querying.getvar(meltwater, 'wfiform', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
melt_M = wfiform_M + wfimelt_M
sfc_salt_flux_restore_M = cc.querying.getvar(meltwater, 'sfc_salt_flux_restore', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
sfc_salt_flux_ice_M = cc.querying.getvar(meltwater, 'sfc_salt_flux_ice', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
fprec_M = cc.querying.getvar(meltwater, 'fprec', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
lprec_M = cc.querying.getvar(meltwater, 'lprec', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
evap_M = cc.querying.getvar(meltwater, 'evap', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,:,:].sel(yt_ocean=lat_slice)
```

```python
SSS_ctrl = cc.querying.getvar(control, 'salt', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,0,:,:].sel(yt_ocean=lat_slice)
SSS_melt = cc.querying.getvar(meltwater, 'salt', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:120,0,:,:].sel(yt_ocean=lat_slice)
```

```python
# convert salt fluxes to FW equivalent:

# A. Kiss don't think we should add this salt fluxes here; we're comparing them but the surface FW budget is given by:
# PME_RIVER = RIVER + MELT (wfiform+wfimelt) + PME (pme_net - (wfiform+wfimelt)) 
FW_restore_C = -(sfc_salt_flux_restore_C*1000)/SSS_ctrl
FW_restore_M = -(sfc_salt_flux_restore_M*1000)/SSS_melt

# He also says we should use "the fixed bulk salinity of sea ice (ice_salt_concentration = 0.005 kg salt / kg ice), not SSS"
SSSi = 1/0.005

FW_ice_C = -(sfc_salt_flux_ice_C*1000)/SSSi
FW_ice_M = -(sfc_salt_flux_ice_M*1000)/SSSi
```

```python
# topography data for plotting:
ht = cc.querying.getvar(control,'ht' , master_session,n=1).sel(yt_ocean=lat_slice)
land_mask = np.squeeze(ht.values)*0
land_mask[np.isnan(land_mask)] = 1
land_mask = np.where(land_mask==1,land_mask,np.nan)
land_mask_masked = np.ma.masked_where((land_mask==0),land_mask)
# make land go all the way to -90S:
land_mask_lat = ht.yt_ocean.values
land_mask_lat[0] = -90

fontsize=15
```

```python
area_old = xr.open_dataset('/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/output012/ocean/ocean_grid.nc')['area_t']
varr = xr.open_dataset('/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/output012/ocean/ocean.nc')['salt']
area_corr = area_old*(varr.isel(time=0,st_ocean=0)*0 +1)
```

```python
%%time
import numpy.ma as ma

area_t = area_corr.sel(yt_ocean=lat_slice)


# function to do annual avg and horiz. integral: ONLY SHELF
def var_tavg_hint_shelf(var2d):
    var = var2d.mean('time') # annual mean
    var_masked = ma.masked_where(ht > 1000, var) # masked offshore 1000m isobath
    var_masked_area = var_masked*area_t # multiply by area
    var_hint = np.nansum(np.nansum(var_masked_area, axis=1), axis=0) # sum along x/y axis

    return var_hint

# function to do annual avg and horiz. integral: GLOBAL
def var_tavg_hint(var2d):
    var = var2d.mean('time') # annual mean
    var_area = var*area_t # multiply by area
    var_hint = np.nansum(np.nansum(var_area, axis=1), axis=0) # sum along x/y axis

    return var_hint
```

```python
%%time

pme_net_annual_hint_shelf_ctrl = var_tavg_hint_shelf(pme_net_C)
pme_river_annual_hint_shelf_ctrl = var_tavg_hint_shelf(pme_river_C)
river_annual_hint_shelf_ctrl = var_tavg_hint_shelf(river_C)
melt_annual_hint_shelf_ctrl = var_tavg_hint_shelf(melt_C)
FW_restore_annual_hint_shelf_ctrl = var_tavg_hint_shelf(FW_restore_C)
FW_ice_annual_hint_shelf_ctrl = var_tavg_hint_shelf(FW_ice_C)
```

```python
%%time

pme_net_annual_hint_shelf_M = var_tavg_hint_shelf(pme_net_M)
pme_river_annual_hint_shelf_M = var_tavg_hint_shelf(pme_river_M)
river_annual_hint_shelf_M = var_tavg_hint_shelf(river_M)
melt_annual_hint_shelf_M = var_tavg_hint_shelf(melt_M)
FW_restore_annual_hint_shelf_M = var_tavg_hint_shelf(FW_restore_M)
FW_ice_annual_hint_shelf_M = var_tavg_hint_shelf(FW_ice_M)
```

```python

%%time
## do also for melt components: wfiform/wfimelt

wfimelt_annual_hint_shelf_ctrl = var_tavg_hint_shelf(wfimelt_C)
wfiform_annual_hint_shelf_ctrl = var_tavg_hint_shelf(wfiform_C)
wfimelt_annual_hint_shelf_M = var_tavg_hint_shelf(wfimelt_M)
wfiform_annual_hint_shelf_M = var_tavg_hint_shelf(wfiform_M)
```
