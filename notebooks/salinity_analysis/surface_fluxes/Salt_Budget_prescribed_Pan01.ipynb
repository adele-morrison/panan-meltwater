```python
# Shelf mask

def shelf_mask_isobath(var, model_dict):

    paths = {
             "mom5": "/g/data/ik11/grids/Antarctic_slope_contour_1000m.npz",
             "mom6_01": "/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_01deg.nc",
            "mom6_005": "/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_005deg.nc"
             }

    var = var.cf.sel({'latitude': slice(-90, -59)})

    if paths[model_dict][-3:] == '.nc':
        shelf_mask = xr.open_dataset(paths[model_dict])['contour_masked_above']
    else:
        contour_file = np.load(paths[model_dict])
        shelf_mask = xr.DataArray(contour_file['contour_masked_above'],
                                  coords = var.coords, 
                                  dims = var.dims,
                                  name = 'contour_masked_above')
    
    shelf_mask = xr.where(shelf_mask == 0, 1, 0)
    masked_var = var * shelf_mask
    
    return masked_var, shelf_mask
```

### Salinity budget

For the salt/FW fluxes I have the following variables:
- `wfo` = Water Flux Into Sea Water
   - Net surface water flux into the ocean, including precipitation, evaporation, river runoff, and sea ice melt/freeze. Positive into the ocean.
- `fsitherm` = water flux to ocean from sea ice melt(> 0) or form(< 0)
- `salt_flux` = Net salt flux into ocean at surface (restoring + sea-ice)
- `salt_flux_added` = Salt flux into ocean at surface due to restoring or flux adjustment
- `precip` = Liquid + frozen precipitation into ocean
- `lrunoff`= Liquid runoff (rivers) into ocean

I am just going to look at the salt fluxes and do the fresh water fluxes separately I think.

Surface budget is given by:

F_{surface} = `salt_flux`

Total budget is:
$$ DS/Dt = F_{surface} +F_{advection} $$

So I am trying to find $F_{advection}$ by doing:
$$ F_{advection} = DS/Dt - F_{surface} $$

```python
sfc_salt_flux_C = cc.querying.getvar(control, 'salt_flux', master_session, start_time=start_time, end_time=end_time, frequency='1 monthly')[0:14,:,:].sel(yh=lat_slice)
net_sfc_salt_C = sfc_salt_flux_C.sel(time=slice(start_time,end_time))
```

```python
import pandas as pd

# assume net_sfc_salt_C.time is a datetime-like coordinate
time = xr.cftime_range(start=str(net_sfc_salt_C.time.values[0]),
                       end=str(net_sfc_salt_C.time.values[-1]),
                       freq='MS')  # if you have monthly means

# --- compute days in each month from actual timestamps ---
days_in_month = xr.DataArray(
    [pd.Period(str(t)).days_in_month for t in net_sfc_salt_C.time.values],
    coords={'time': net_sfc_salt_C.time},
    dims='time'
)

# convert to seconds per month
sec_in_month = days_in_month * 24 * 60 * 60
```

```python
# This will now be in J integrated over the year.
net_sfc_salt_C = ((net_sfc_salt_C*sec_in_month).sum('time')*area).sum().compute()

print('Control Net surface salt input over this period =',net_sfc_salt_C.values/1e20,'x 10^20 kg of salt')
```
