```python
## Creating shelf mask another way

def shelf_mask_isobath(var, model_dict):

    paths = {
             "mom5": "/g/data/ik11/grids/Antarctic_slope_contour_1000m.npz",
             "mom6_01": "/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_01deg.nc",
            "mom6_005": "/g/data/ik11/grids/Antarctic_slope_contour_1000m_MOM6_005deg.nc"
             }

    var = var.cf.sel({'latitude': slice(-90, -59)})

    if paths[model_dict][-3:] == '.nc':
        shelf_mask = xr.open_dataset(paths[model_dict])['contour_masked_above']
    else:
        contour_file = np.load(paths[model_dict])
        shelf_mask = xr.DataArray(contour_file['contour_masked_above'],
                                  coords = var.coords, 
                                  dims = var.dims,
                                  name = 'contour_masked_above')
    
    shelf_mask = xr.where(shelf_mask == 0, 1, 0)
    masked_var = var * shelf_mask
    
    return masked_var, shelf_mask
```

```python
depth = xr.open_dataset('/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/output012/ocean/ocean_grid.nc')['ht']
land_mask = (0 * depth).fillna(1)
depth_shelf, shelf_mask = shelf_mask_isobath(depth, 'mom5')
```

```python
#Time variant variables
def preprocess(ds):
    ds = ds['salt'].sel(yt_ocean=shelf_mask['yt_ocean'])
    return ds

cont = xr.open_mfdataset(pathsC, preprocess = preprocess, chunks = 'auto')
melt = xr.open_mfdataset(pathsM, preprocess = preprocess, chunks = 'auto')
```

```python
#Time slicing
cont = cont.sel(time=slice('1910-01-01','1919-12-31'))
melt = melt.sel(time=slice('1910-01-01','1919-12-31'))
```

```python
#Time invariant variables
area = xr.open_dataset('/g/data/ik11/outputs/access-om2-01/01deg_jra55v13_ryf9091_21mbath/output012/ocean/ocean_grid.nc')['area_t'].sel(xt_ocean=shelf_mask['xt_ocean'])
```

```python
# Salinity variable
saltC = cont['salt']
saltM = melt['salt']
```

```python
# Correcting area
area_corr = area*(saltC.isel(time=0,st_ocean=0)*0 +1)
```

```python
area_corr.plot()
```

### Bottom mask

```python
bottom_mask_saltC = saltC.where(~np.isfinite(saltC.shift({'st_ocean': -1})))
bottom_mask_saltC = ~np.isnan(bottom_mask_saltC)
```

```python
bottom_mask_saltM = saltM.where(~np.isfinite(saltM.shift({'st_ocean': -1})))
bottom_mask_saltM = ~np.isnan(bottom_mask_saltM)
```

```python
botsalt_cont = saltC.where(bottom_mask_saltC).sum(dim='st_ocean')
botsalt_pert = saltM.where(bottom_mask_saltM).sum(dim='st_ocean')
```

### Time series

```python
# Restrict to shelf
botsalt_cont_shelf = botsalt_cont.where(shelf_mask == 1)
botsalt_pert_shelf = botsalt_pert.where(shelf_mask == 1)
```

```python
botsalt_cont_shelf.isel(time=0).plot()
```

```python
# Removing land [if needed, check above]
botsalt_cont_shelf = botsalt_cont_shelf*(saltC.isel(time=0,st_ocean=0)*0 +1)
botsalt_pert_shelf = botsalt_pert_shelf*(saltM.isel(time=0,st_ocean=0)*0 +1)
```

```python
# Check mask again
botsalt_cont_shelf.isel(time=0).plot()
```

```python
# Check mask again
botsalt_pert_shelf.isel(time=0).plot()
```

```python
# Spatial average
botsalt_cont_time = (area_corr * botsalt_cont_shelf).sum(['xt_ocean', 'yt_ocean']) / area_corr.where(shelf_mask == 1).sum(['xt_ocean', 'yt_ocean'])
botsalt_pert_time = (area_corr * botsalt_pert_shelf).sum(['xt_ocean', 'yt_ocean']) / area_corr.where(shelf_mask == 1).sum(['xt_ocean', 'yt_ocean'])
```

#### Mapping bottom salinity

```python
saltC_map = saltC.sel(time=slice('1915-01-01','1919-12-31'))
saltM_map = saltM.sel(time=slice('1915-01-01','1919-12-31'))
```

```python
botsalt_cont_map = saltC_map.where(bottom_mask_saltC).sum(dim='st_ocean').mean(dim='time')
botsalt_pert_map = saltM_map.where(bottom_mask_saltM).sum(dim='st_ocean').mean(dim='time')
botsalt_anom_map = botsalt_pert_map - botsalt_cont_map
```
